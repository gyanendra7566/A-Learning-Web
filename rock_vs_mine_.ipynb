```python
X = sonar_data.drop(columns = 60)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
model.fit(X_train, Y_train)
```
