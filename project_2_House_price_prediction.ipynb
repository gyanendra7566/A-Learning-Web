Importing the Dependencyies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Prince Data


```python
house_price_data = fetch_california_housing(as_frame=True)
```

```python
print(house_price_data)
```

```python
# Loading the Datasets to Pandas DataFrame
house_price_dataframe = pd.DataFrame(house_price_data.data , columns = house_price_data.feature_names)
```

```python
#Print the Upper 5 rows of Data
house_price_dataframe.head()
```

Add the targe(price) column to the dataframe


```python
house_price_dataframe['price'] = house_price_data.target
```

```python
house_price_dataframe.head()
```

```python
# Checking the number of rows and columns in dataframe
house_price_dataframe.shape
```

```python
# Check for missing values
house_price_dataframe.isnull().sum()
```

```python
#Statistical measure of the dataset
house_price_dataframe.describe()
```

Understanding the Correlation between various data

1. positive correlation.
2. Negative correlation.

```python
correlation_data = house_price_dataframe.corr()
```

```python
# Constructing the heatmap to understan the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation_data , cbar=True, square=True, fmt='.1f', annot=True, annot_kws = {'size':8}, cmap='Blues')
```

```python
#Spliting the data and Target
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Spliting the data into Training data and Testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state =2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGboost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

Evaluation

prediction of training data

```python
# accuracy of prediction of training data
model_data_prediction = model.predict(X_train)
```

```python
print(model_data_prediction)
```

```python
# 1. R square error

score_1 = metrics.r2_score(Y_train, model_data_prediction)

# 2. mean square error
score_2 = metrics.mean_absolute_error(Y_train, model_data_prediction)

print("R score error: ",score_1)
print("Mean score error: ",score_2)


```

Visualization of Actual prices vs predicted prices

```python
plt.scatter = (Y_train, model_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("predicted prices")
plt.title("Actual prices vs predicted prices")
plt.show()
```

prediction of test data

```python
model_test_data = model.predict(X_test)
```

```python
#1. R square error
score_1 = metrics.r2_score(Y_test, model_test_data)

# 2. mean square error
score_2 = metrics.mean_absolute_error(Y_test, model_test_data)

print("R score error: ",score_1)
print("Mean score error: ",score_2)
```
